```python
import torch
from fastai.data.all import *
from transforms import *
```

```python
sample_rate = 22050
n_fft = 1024
hop_length = 256
path = Path("../data/LJSpeech-1.1/wavs")
```

```python
mel_tfms = [Load_audio(sample_rate), Mel_transform(sample_rate, n_fft, hop_length)]
energy_tfms = [Load_audio(sample_rate), Stft_transform(n_fft, hop_length), 
              extract_amplitude, l2_norm]
pitch_tfms = [Load_audio(sample_rate), Extract_pitch_contour(sample_rate, hop_length)]
duration_tfms = [Change_extension(".TextGrid"), load_text_grid, 
                 Get_durations(sample_rate, hop_length)]
phones_tfms = [Change_extension(".TextGrid"), load_text_grid,
               get_phonemes]
```

```python
pad_size = 800
sequence_pad_tfms = [Pad_sequence(pad_size)]
mel_pad_tfms = [Pad_mel(0, pad_size)]
```

```python
get_audio = FileGetter(extensions=".wav")
```

```python
files = get_audio(path)
```

```python
file = files[0]

mel = Pipeline(mel_tfms + [db_transform] + mel_pad_tfms)(file)
energy = Pipeline(energy_tfms + sequence_pad_tfms)(file)
pitch = Pipeline(pitch_tfms + sequence_pad_tfms)(file)
durations = Pipeline(duration_tfms)(file)
phones = Pipeline(phones_tfms)(file)
```

```python
dataset_size = 0.3
```

```python
tfms = [phones_tfms, duration_tfms, mel_tfms + mel_pad_tfms, 
        energy_tfms + sequence_pad_tfms,pitch_tfms + sequence_pad_tfms]
files = get_audio(path)[:int(len(files)*dataset_size)]
splits = RandomSplitter()(files)
```

```python
dsets = Datasets(files, tfms, splits=splits)
```
